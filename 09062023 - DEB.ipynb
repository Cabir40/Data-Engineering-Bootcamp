```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('mydata/Mall_Customers.csv')
customer_data.head()
```

```python
customer_data.shape
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize = (10,7))
plt.title("Customer Dendrograms")
dend = shc.dendrogram(shc.linkage(data, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize = (5,5))
plt.scatter(data[:, 0], data[:, 1], c = cluster.labels_, cmap = 'rainbow')
```

```python
#Principal Component Analysis (Temel Bileşenler Analizi)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
df.head()
```

```python
df.shape
```

```python
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
scaled_data
```

```python
pca = PCA(n_components = 3)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
print("Before PCA shape", df.shape)
print("After PCA shape", x_pca.shape)
```

```python
import matplotlib.pyplot as plt
pca = PCA(random_state=7)
pca.fit(scaled_data)

var_cumu = np.cumsum(pca.explained_variance_ratio_)*100
k = np.argmax(var_cumu>90)

plt.figure(figsize=(10,5))
plt.axvline(x=k, color='red',linestyle='--')
plt.axhline(y=90,color='blue',linestyle='--')
plt.grid()
ax = plt.plot(var_cumu)
```

```python
plt.figure(figsize = (6,6))

plt.scatter(x_pca[:, 0], x_pca[:, 1], c = cancer['target'])

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#Ensemble Methods - Random Forest
```

```python
dataset = pd.read_csv('mydata/Iris.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 50)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report")
print(result1)
result2 = accuracy_score(y_test, y_pred)
print("Accuracy:", result2)
```
